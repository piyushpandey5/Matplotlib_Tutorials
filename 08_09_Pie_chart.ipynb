```python
classes=["Python", "R" ,"Machine Learning", " Artificial Intelligence", "Data Science" ]
class1_students = [45,15,35,25,30]
explode=[0.3,0,0,0.2,0]
color=["c", "b", "r", "y", "g"]
textprops={"fontsize":15}
plt.pie(class1_students,labels=classes, explode=explode,colors = color,autopct="%0.3f%%"
        ,shadow=True,radius=1.4,startangle=270,textprops = textprops)
plt.show()
```

```python
plt.figure(figsize=(3,2))
wedgeprops={"linewidth":4,"width":3,"edgecolor":"k"}
plt.pie(
    class1_students,
    explode=explode,
    labels=classes,
    colors=color,
    autopct="0.2f %%",
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.7,
    startangle=270,
    radius=1,
    counterclock=True,
    wedgeprops=wedgeprops,
    textprops=textprops,
    center=(2, 3),
    frame=True,
    rotatelabels=False,
    
    normalize=None,
    data=None,
)
plt.show()
```

```python
explode=[0.3,0,0,0.2,0]
color=["c", "b", "r", "y", "g"]
textprops={"fontsize":15}
plt.pie(class1_students,labels=classes, explode=explode,colors = color,autopct="%0.3f%%"
        ,shadow=True,radius=1.4,startangle=270,textprops = textprops)
plt.legend(loc=3)
plt.show()
```
