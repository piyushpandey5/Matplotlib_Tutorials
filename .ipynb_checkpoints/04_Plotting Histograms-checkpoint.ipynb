```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45,(100))
py_student_age = np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age) #make histogram
plt.title("ML student age histograms")
plt.xlabel("Student age category")
plt.ylabel("No of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",
         orientation="horizontal",color="m",label="ml student") #make histogram  , rwidth changes the width of the bars,
        #orientation changes from orientation of graph from horizontal to vertical
plt.title("ML student age histograms")
plt.xlabel("Student age category")
plt.ylabel("No of students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",
         orientation="vertical",color="m",label="ml student") #make histogram  , rwidth changes the width of the bars,
        #orientation changes from orientation of graph from horizontal to vertical
plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",
         orientation="vertical",color="y",label="Py student")
plt.title("ML student age histograms")
plt.xlabel("Student age category")
plt.ylabel("No of students age")
plt.legend()
plt.show()
```
