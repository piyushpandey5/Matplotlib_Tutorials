```python
df_google_play_store_apps=pd.read_csv("F:\\ML Datasets\\Google play store datasets\\googleplaystore.csv",nrows=1000)
df_google_play_store_apps
```

```python

df_google_play_store_apps
```

```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google play Store Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c = "r", marker = "*", s =100,alpha = 0.2,linewidths = 10,edgecolor="g") # s is for size  #alpha is for transperancy 
plt.title("Google play Store Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c = "r", marker = "*", s =100,alpha = 0.5,linewidths = 10,edgecolor="g")
plt.scatter(x,df_google_play_store_apps["Installs"], c = "y", marker = "<", s =100,alpha = 0.5,linewidths = 10,edgecolor="k")

plt.title("Google play Store Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Installs and Reviews")
```
