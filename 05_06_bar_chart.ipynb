# Ploting Bar Chart


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["Python","R","AI","ML","DS"]
class1_students=[30,10,20,25,10] #out of 100 student in each class
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students) #barh will plot the graph horizontaly
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color="y",edgecolor="m",linewidth=5,alpha=0.5,linestyle="--",label=" Class 1 student",visible=True)
#edgecolor changes the color of the edge of the graph, alpha maintains the capacity of color, visible enables or disables the visibility of the graph
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.6,align="edge",color="black"
        ,edgecolor="m",linewidth=5,linestyle="--",label=" Class 1 student",visible=True)
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.barh(classes,class1_students,align="edge",color="black"
        ,edgecolor="m",linewidth=5,linestyle="--",label=" Class 1 student",visible=True)
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.6,align="edge",color="k"
        ,edgecolor="m",linewidth=5,linestyle="--",label=" Class 1 student",visible=True)
plt.title("bar chart of IAIP Class",fontsize=15)
plt.xlabel("Classes",fontsize=13)
plt.ylabel("No.of student",fontsize=13)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.2,align="edge",color="b"
        ,edgecolor="m",linewidth=5,linestyle="--",label=" Class 1 student",visible=True)
plt.bar(classes,class2_students,width=0.2,align="edge",color="g"
        ,edgecolor="y",linewidth=5,linestyle="--",label=" Class 1 student",visible=True)
plt.title("bar chart of IAIP Class",fontsize=15)
plt.xlabel("Classes",fontsize=13)
plt.ylabel("No.of student",fontsize=13)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color="b"
        ,label=" Class 1 student",visible=True)
plt.bar(classes_index + width, class2_students,width,color="g"
        ,label=" Class 2 student",visible=True) #class_index + width add the width of the graph to make another graph in its side
plt.bar(classes_index + width + width, class3_students,width,color="y"
        ,label=" Class 3 student",visible=True)
plt.xticks(classes_index + width +width, classes,rotation=20) #replace the numbering of 0 1 2 with the name of the classes and rotation change the angle of the text
plt.title("Bar chart of IAIP Class",fontsize=15)
plt.xlabel("Classes",fontsize=13)
plt.ylabel("No.of student",fontsize=13)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.barh(classes_index,class1_students,width,color="b"
        ,label=" Class 1 student",visible=True)
plt.barh(classes_index + width, class2_students,width,color="g"
        ,label=" Class 2 student",visible=True) #class_index + width add the width of the graph to make another graph in its side
plt.barh(classes_index + width + width, class3_students,width,color="y"
        ,label=" Class 3 student",visible=True)
plt.yticks(classes_index + width +width, classes,rotation=20) #replace the numbering of 0 1 2 with the name of the classes and rotation change the angle of the text
plt.title("Bar chart of IAIP Class",fontsize=15)
plt.ylabel("Classes",fontsize=13)
plt.xlabel("No.of student",fontsize=13)
plt.legend()
plt.show()
```
